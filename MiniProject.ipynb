```python
def convert_range_to_midpoint(value):
    if isinstance(value, str) and '+' in value:
        # Handle the range '120+' by choosing a large integer value
        return 1e10  # Choose a suitable large integer
    else:
        # For other values, convert them to integers
        return int(value)

# Apply the function to the 'Days_spent' column
data_clean['Days_spent'] = data_clean['Days_spent'].apply(convert_range_to_midpoint)
```

```python
#Splitting data into x and y (x:not Dependable variable, y:Dependable variable)
x=data_clean.drop(['claim'],axis=1)
y=data_clean['claim']
```

```python
#importing sklearn for training and testing data
from sklearn.model_selection import train_test_split
```

```python
#Assigning attributes for train test splitting (Test data:20% of total data)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
#Importing Regression ML algorithms and models from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
#Creating instances of the algorithm
LR=LinearRegression()
LR.fit(x_train,y_train)

svr=SVR()
svr.fit(x_train,y_train)

RF=RandomForestRegressor()
RF.fit(x_train,y_train)

GR=GradientBoostingRegressor()
GR.fit(x_train,y_train)
```

```python
#Comparing results of training with test data and prediction is stored in variables
y_pred1=LR.predict(x_test)
y_pred2=RF.predict(x_test)
y_pred3=RF.predict(x_test)
y_pred4=GR.predict(x_test)
df1=pd.DataFrame({'Actual':y_test,'LR':y_pred1,'Sum':y_pred2,'RF':y_pred3,'GR':y_pred4})
print(df1)
```

```python
#Graphical plotting of each regression model with its actual and predicted value competing eachothers
import matplotlib.pyplot as plt
plt.subplot(221)
plt.plot(df1['Actual'].iloc[0:11],label='Actual')
plt.plot(df1['LR'].iloc[0:11],label='LR')
plt.legend()

plt.subplot(222)
plt.plot(df1['Actual'].iloc[0:11],label='Actual')
plt.plot(df1['Sum'].iloc[0:11],label='Sum')
plt.legend()

plt.subplot(223)
plt.plot(df1['Actual'].iloc[0:11],label='Actual')
plt.plot(df1['RF'].iloc[0:11],label='RF')
plt.legend()

plt.subplot(224)
plt.plot(df1['Actual'].iloc[0:11],label='Actual')
plt.plot(df1['GR'].iloc[0:11],label='GR')
plt.legend()
```

```python
#Importing metric system r2 score
from sklearn import metrics
```

```python
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
print(score1,score2,score3,score4)
```
